## 3. Run Stock Market Investment Optimizer V1

### a) Run without constrains

```python
import warnings
warnings.filterwarnings("ignore")

sm = StockMarketInvestor(sp_hist, 0.01)
sm.train()
ret = sm.optimize()
sm.evaluate(ret, sp_hist, 5000)
```

### b) Run with constrains include i.e., include Apple in my portfolio

```python
sm = StockMarketInvestor(sp_hist, 0.01, include=['AAPL'])
sm.train()
ret = sm.optimize()
sm.evaluate(ret, sp_hist, 5000)
```

### c) Run with constrains include Aaple and exclude other stocks

```python
sm = StockMarketInvestor(sp_hist, 0.05, include=['AAPL'], exclude=['TSLA','FCX'])
sm.train()
ret = sm.optimize()
sm.evaluate(ret, sp_hist, 5000)
```

## 4.Merge and Prepare data for individual stock assessment

```python
ret = ret.merge(sp_comp, left_on='Symbol', right_on='Symbol')
ret = ret.sort_values(by="InvestmentPercentage", ascending=False)
#get more information about selected stocks, like volume
top_stocks = list(ret.sort_values(by="InvestmentPercentage", ascending=False)["Symbol"][:4])
print(top_stocks)
top_stocks_dict = Loader.get_selected_ticker_historical(top_stocks)
```

## 5. Machine Learning on Individual stock

## Feature Engineering

```python
keys = top_stocks_dict.keys()
for i in keys:
    #call loader function to create features and training data from ticker historical and volume data. 
    top_stocks_dict[i] = Loader.create_features_training(top_stocks_dict[i])
```

## Test Train Split

```python
mos = top_stocks_dict['NVDA']
split = int(0.85 * mos.shape[0])
X_train, y_train = mos.iloc[:split, 1:],  mos.iloc[:split, :1]
X_test, y_test = mos.iloc[split:, 1:], mos.iloc[split:, :1] 
```

```python
print(X_train.shape, y_train.shape,X_test.shape,y_test.shape,)
```

## Prediction using Random Forest Regressor

```python
clf = RandomForestRegressor_StockPredictor()
clf.fit_predict(mos, X_train, y_train, X_test, y_test)
```
