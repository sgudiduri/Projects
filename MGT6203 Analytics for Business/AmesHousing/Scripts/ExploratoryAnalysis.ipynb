```python

# Getting correlations
corr = num_fields.corr()
corr_desc = (corr['SalePrice'].sort_values(ascending=False))
# corr_desc.drop(['SalePrice'])
type(corr_desc)
corr_lst = (corr_desc.index.tolist())
# High_corr_Features = pd.DataFrame({'Feature':corr_desc.index,'corr':corr_desc.values})
#selecting top 15 values with high correlation
# High_corr_Features.head()

High_corr = corr_lst[1:16]
print(High_corr)
             
 
```

```python
# df.head()
for i in High_corr:
    data=df1.copy

    df1.groupby(i)['SalePrice'].median().plot()
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.show()
```
