```python
#Missing Values in numeric columns


missing_val_num_col = [i for i in FE_df.columns if FE_df[i].isnull().sum()>1 and FE_df[i].dtypes!="O"]
for i in missing_val_num_col:
    print('missing_val_num_col {}{}%'.format(i, np.round(FE_df[i].isnull().mean(),4)))
```

As seen in pandas profiling report, the distribution for LotFrontage and MasVnrArea is skewed, we cannot fill the missing values with mean and hence we will use median


```python
import math
for i in missing_val_num_col:
    median_val = FE_df[i].median()
#     median_val = ceil(median_val)
    #print(median_val)
    FE_df[i].fillna(median_val, inplace=True)
```

```python
#checking missing numerical val cols
for i in missing_val_num_col:
    print('missing_val_num_col {}{}%'.format(i, np.round(FE_df[i].isnull().mean(),4)))
```
